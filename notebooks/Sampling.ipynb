```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
n = 10000
usamples = np.random.random(n)
#usamples

mu  = 0
sig = 1.0
# norm.ppf (percent point function) is the inverse of the CDF
gsamples = stats.norm.ppf(usamples, mu, sig)
```

```python
a,b,c = plt.hist(gsamples, bins=np.arange(-5,5,0.2))
```
