# Gaussian (normal) distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

A gaussian distribution is characterized by the mean $\mu$ and standard deviation $\sigma$.  See:

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html

```python
# mean = 2, sigma = 1
mu  = 2.0
sig = 1.0
x = np.arange(-5.0, 7.1, 0.1)
p = stats.norm.pdf(x, loc=mu, scale=sig)
#print(x, p)
```

```python
plt.plot(x, p)
plt.xlabel('$x$')
plt.ylabel('$p_G(x|\mu, \sigma)$')
```

As with other PDFs, you can sample $x$ values from a gaussian PDF with `np.random.normal`.

```python
np.random.seed(12345)

# Let's sample from a Gaussian distribution with mean=0, sigma=1
mean = 0.0
std  = 1.0
d1 = np.random.normal(loc=mean, scale=std, size=1000000)

# Create bins of x values
nx = np.arange(-5.0, 5.1, 0.1)

# Let's plot the normalized histogram of the samples
n1, bins1, patches1 = plt.hist(d1, bins=nx, density=True)

# and overplot the analytical Gaussian PDF.
ny = stats.norm.pdf(nx, mean, std)
plt.plot(nx, ny, color='r')
```

Now let's see what happens with we sample from two different gaussians ($\mu_1$, $\sigma_1$) and ($\mu_2$, $\sigma_2$) and plot the distribution of the difference $\mu_1-\mu_2$.  You can see that the result is a new gaussian with a mean of $\mu_1-\mu_2$ and standard deviation of $\sqrt{\sigma_1^2+\sigma_2^2}$.

```python
np.random.seed(12345)

mean1 = 2.0
std1  = 1.0
d1 = np.random.normal(loc=mean1, scale=std1, size=1000000)

mean2 = 1.0
std2  = 2.0
d2 = np.random.normal(loc=mean2, scale=std2, size=1000000)

d3 = d1-d2

nx = np.arange(-10.0, 10.1, 0.1)
ny = stats.norm.pdf(nx, mean1-mean2, np.sqrt(std1**2 + std2**2))
n1, bins1, patches1 = plt.hist(d3,
                               bins=nx,
                               density=True)
plt.plot(nx, ny, color='r')
```
