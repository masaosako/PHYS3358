```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

Let simulate a series of 3 coin flips for values of H = probability of heads, and count the fraction of times that end up with 2 heads and 1 tail.  This is just another way of validating the posterior PDF derived from Bayes' Theorem.

```python
nsets = 10000

n, k = 3, 2

hval = np.arange(0, 1.05, 0.05)
pval = np.empty_like(hval)

for ih, h in enumerate(hval):
    nmatch = 0
    for i in range(nsets):
        # flipsim is an array that is 0 tails or 1 heads
        flipsim = np.where(np.random.random(n)>=h, 0, 1)
        nheads = flipsim.sum()
        if np.isclose(nheads, k, rtol=1.e-2)==True:
            nmatch += 1

    pval[ih] = nmatch/nsets
```

```python
plt.plot(hval, pval, label="simulation")
y3 = binom.pmf(2, 3, hval)
plt.plot(hval, y3, label="prediction")
plt.legend()
```
