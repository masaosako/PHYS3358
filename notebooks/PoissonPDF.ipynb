# **Poisson distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

Poisson distribution is characterized by a single parameter $\lambda$.  See:

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html

```python
# expectation of lam=1 event on average
lam = 1
x = np.arange(11)   # = [0 1 .. 10]
p = stats.poisson.pmf(x, lam)
# Here are the probabilities of observing 0, 1, .., 10 events
print(x, p)
```

```python
plt.bar(x, p, width=0.3)
```

Below is the probability of observing 9 events if the expectation is 2.315e-5 (neutrinos from SN1987A).

```python
stats.poisson.pmf(9, 2.315e-5)
```
