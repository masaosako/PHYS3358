**Question 1**:

(4 points) For the uniform distribution $f_U (x|a, b)$ over the interval $a$ to $b$,
$$
f_U (x|a, b) = 
\begin{cases}
1/(b âˆ’ a) & \text{for}~ a \le x \le b \\
0 & \text{elsewhere}
\end{cases}
$$
calculate the expressions for the mean $\mu$, variance $\sigma^2$, skewness $\Sigma$, and kurtosis $K$ in terms
of $a$ and $b$ 

(HINT: the answers are in Section 5.8.2 of Gregory).


### BEGIN SOLUTION

$$
\begin{align}
\mu &= \langle x \rangle \\
&= \int x f_U(x) dx\\
&= \int_a^b \frac{x}{b-a} dx\\
&= \frac{1}{b-a} \left(\frac{b^2}{2} - \frac{a^2}{2}\right) \\
&= \boxed{\frac{a+b}{2}}
\end{align}
$$

---

$$
\begin{align}
\sigma^2 &= \langle (x-\mu)^2 \rangle \\
&= \int (x-\mu)^2 f_U(x) dx\\
&= \int_a^b (x-\mu)^2 \frac{1}{b-a} dx\\
&= \frac{1}{b-a} \left(\frac{(b-\mu)^3}{3} - \frac{(a-\mu)^3}{3}\right) \\
&= \frac{1}{b-a} \left(\frac{\left(\frac{b-a}{2}\right)^3}{3} - \frac{\left(-\frac{b-a}{2}\right)^3}{3}\right) \\
&= \frac{1}{b-a} \left(2\frac{(b-a)^3}{24}\right) \\
&= \boxed{\frac{(b-a)^2}{12}}
\end{align}
$$

---

$$
\begin{align}
\Sigma &= \frac{1}{\sigma^3} \langle (x-\mu)^3 \rangle \\
&= \frac{1}{\sigma^3} \int (x-\mu)^3 f_U(x) dx\\
&= \frac{1}{\sigma^3} \int_a^b (x-\mu)^3 \frac{1}{b-a} dx\\
&= \frac{1}{b-a} \left(\frac{(b-\mu)^4}{4} - \frac{(a-\mu)^4}{4}\right) \\
&= \frac{1}{b-a} \left(\frac{\left(\frac{b-a}{2}\right)^4}{4} - \frac{\left(-\frac{b-a}{2}\right)^4}{4}\right) \\
&= \boxed{0}
\end{align}
$$

---

$$
\begin{align}
K &= \frac{1}{\sigma^4} \langle (x-\mu)^4 \rangle - 3\\
&= \left( \frac{1}{\sigma^4} \int (x-\mu)^4 f_U(x) dx \right) - 3\\
&= \frac{1}{\sigma^4} \left(\int_b^a (x-\mu)^4 \frac{1}{b-a} dx \right) - 3\\
&= \frac{12^2}{(b-a)^5} \left(\frac{(b-\mu)^5}{5} - \frac{(a-\mu)^5}{5}\right) -3\\
&= \frac{144}{(b-a)^5} \left(\frac{\left(\frac{b-a}{2}\right)^5}{5} - \frac{\left(-\frac{b-a}{2}\right)^5}{5}\right) -3\\
&= \frac{144}{(b-a)^5} \left(\frac{2(b-a)^5}{160}\right) -3\\
&= \frac{9}{5} - 3 \\
&= \boxed{-\frac{6}{5}}
\end{align}
$$

### END SOLUTION

**Question 2**:
    
(4 pts) For the binomial distribution,
  \begin{equation}
    p_B(x| n, p) = \binom{n}{x}\,p^x\,q^{n-x} = \frac{n!}{x!\,(n-x)!}\,p^x\,(1-p)^{n-x}
  \end{equation}
  for observing $x$ of $n$ items and $p$ and $q=(1-p)$ are the success
  and failure probabilities, respectively, 
  using your favorite method, calculate in terms of $n$, $p$, and $q$ the expressions for:
  
a) (4 pts) the mean $\mu$ and variance $\sigma^2$
    
b) **EXTRA CREDIT:** (1 pt) the skewness $\Sigma$
    
c) **EXTRA CREDIT:** (2 pts) the kurtosis $K$

(HINT: some of the answers are in Section 5.7.1 of Gregory)

Question 2a:

Hint: It can be helpful to use the fact that the sum of the probabilities over all possibilities $x$ for a
specific choice of $n$ is 1.

### BEGIN SOLUTION

$$
\begin{align}
~ \mu &= \langle x \rangle\\ 
&= \sum_{x=0}^n x p_B(x|n,p)\\
&= \sum_{x=1}^n \frac{x n!}{x! (n-x)!} p^x q^{n-x}\\
&= \sum_{x=1}^n \frac{n!}{(x-1)! (n-x)!} p^x q^{n-x}\\
&= np \sum_{x=1}^n \frac{(n-1)!}{(x-1)! (n-x)!}p^{x-1} q^{n-x}\\
&= np \sum_{x=1}^n p_B(x-1| n-1,p)\\
&= \boxed{np}
\end{align}
$$

---

$$
\begin{align}
~ \sigma^2 &= \langle x^2 \rangle - \mu^2 \\
&= \left( \sum_{x=0}^n x^2 p_B(x|n,p) \right) - \mu^2 \\
&= \left( \sum_{x=1}^n x^2 \frac{n!}{x!(n-x)!} p^x q^{n-x} \right) - \mu^2 \\
&= \left( np \sum_{x=1}^n x \frac{(n-1)!}{(x-1)!(n-x)!} p^{x-1} q^{n-x} \right) - \mu^2 \\
&= \left( np \sum_{x=1}^n \frac{(n-1)!}{(x-1)!(n-x)!} p^{x-1} q^{n-x} + np \sum_{x=1}^n (x-1) \frac{(n-1)!}{(x-1)!(n-x)!} p^{x-1} q^{n-x} \right) -\mu^2 \\
&= \left( np \sum_{x=1}^n p_B(x-1| n-1, p) + np \sum_{x=2}^n \frac{(n-1)!}{(x-2)!(n-x)!} p^{x-1} q^{n-x} \right) -\mu^2 \\
&= \left( np + np ~(n-1)p \sum_{x=2}^n \frac{(n-2)!}{(x-2)!(n-x)!} p^{x-2} q^{n-x} \right) -\mu^2 \\
&= \left( np + n(n-1)p^2 \sum_{x=2}^n p_B(x-2| n-2, p) \right) - \mu^2 \\
&= np + n(n-1)p^2 - (np)^2\\
&= np(1-p)\\
&= \boxed{npq}
\end{align}
$$

### END SOLUTION

Question 2b:

### BEGIN SOLUTION

$$
\begin{align}
~ \Sigma &= \frac{1}{\sigma^3} \langle (x-\mu)^3 \rangle \\
&= \frac{1}{\sigma^3}\left(\langle x^3 \rangle - 3 \langle x^2 \rangle \mu + 3 \langle x \rangle \mu^2 - \mu^3\right) \\
&= \frac{1}{\sigma^3}\left(\langle x^3 \rangle - 3 (\sigma^2 + \mu^2) \mu + 3 \mu \mu^2 - \mu^3\right) \\
&= \frac{1}{\sigma^3}\left(\langle x^3 \rangle - 3 \sigma^2 \mu - \mu^3\right) \\
&= \frac{1}{\sigma^3}\left( \sum_{x=0}^n x^3 p_B(x|n,p) \right) - \frac{3 \sigma^2 \mu + \mu^3}{\sigma^3} \\
&= \frac{1}{\sigma^3}\left( \sum_{x=0}^n x^3 \frac{n!}{x!(n-x)!} p^x q^{n-x} \right) - \frac{3 \sigma^2 \mu + \mu^3}{\sigma^3} \\
&= \frac{1}{\sigma^3}\left[ \sum_{x=1}^n x^2 \frac{n!}{(x-1)!(n-x)!} p^x q^{n-x} \right] - \frac{3 \sigma^2 \mu + \mu^3}{\sigma^3} \\
\end{align}
$$

Split up the term in brackets into three sums using $x^2 = 1 + 3(x-1) + (x-1)(x-2)$.

$$
\begin{align}
\sum_{x=1}^n \frac{n!}{(x-1)!(n-x)!} p^x q^{n-x} &= np \sum_{x=1}^n \frac{(n-1)!}{(x-1)!(n-x)!} p^{x-1} q^{n-x} \\
&= np \sum_{x=1}^n p_B(x-1| n-1, p) \\
&= np
\end{align}
$$

$$
\begin{align}
\sum_{x=1}^n 3(x-1) \frac{n!}{(x-1)!(n-x)!} p^x q^{n-x} &= 3n(n-1)p^2 \sum_{x=2}^n \frac{(n-2)!}{(x-2)!(n-x)!} p^{x-2} q^{n-x} \\
&= 3n(n-1)p^2 \sum_{x=2}^n p_B(x-2| n-2, p) \\
&= 3n(n-1)p^2
\end{align}
$$

$$
\begin{align}
\sum_{x=1}^n (x-1)(x-2) \frac{n!}{(x-1)!(n-x)!} p^x q^{n-x} &= n(n-1)(n-2)p^3 \sum_{x=3}^n \frac{(n-3)!}{(x-3)!(n-x)!} p^{x-3} q^{n-x} \\
&= n(n-1)(n-2)p^3 \sum_{x=3}^n p_B(x-3| n-3, p) \\
&= n(n-1)(n-2)p^3
\end{align}
$$

Putting these back into our original expression, we find

$$
\begin{align}
~ \Sigma &= \frac{1}{\sigma^3}\left[ np + 3n(n-1)p^2 + n(n-1)(n-2)p^3 \right] - \frac{3 \sigma^2 \mu + \mu^3}{\sigma^3} \\
&= \frac{np + 3n(n-1)p^2 + n(n-1)(n-2)p^3 - 3 n^2p^2(1-p) - n^3p^3}{\sigma^3} \\
&= \frac{np(1-p)(1-2p)}{(np(1-p))^{3/2}} \\
&= \frac{1-2p}{(np(1-p))^{1/2}} \\
&= \boxed{\frac{q-p}{\sqrt{npq}}}
\end{align}
$$

### END SOLUTION

Question 2c:

### BEGIN SOLUTION

$$
\begin{align}
~ K &= \frac{1}{\sigma^4} \langle (x-\mu)^4 \rangle -3\\
&= \frac{1}{\sigma^4}\left(\langle x^4 \rangle - 4 \langle x^3 \rangle \mu + 6 \langle x^2 \rangle \mu^2 - 4\langle x\rangle \mu^3 + \mu^4\right) -3\\
&= \frac{1}{\sigma^4}\left(\langle x^4 \rangle - 4 (\Sigma \sigma^3 + 3\sigma^2\mu +\mu^3) \mu + 6 (\sigma^2+\mu^2)\mu^2 - 4 \mu \mu^3 + \mu^4\right) -3\\
&= \frac{1}{\sigma^4}\left(\langle x^4 \rangle - 4 \Sigma \sigma^3 \mu - 6 \sigma^2 \mu^2 - \mu^4\right) -3\\
&= \frac{1}{\sigma^4}\left( \sum_{x=0}^n x^4 p_B(x|n,p) \right) - \frac{4\Sigma \sigma^3 \mu + 6\sigma^2 \mu^2 + \mu^4}{\sigma^4} -3\\
&= \frac{1}{\sigma^4}\left( \sum_{x=0}^n x^4 \frac{n!}{x!(n-x)!} p^x q^{n-x} \right) - \frac{4\Sigma \sigma^3 \mu + 6\sigma^2 \mu^2 + \mu^4 + 3\sigma^4}{\sigma^4}\\
&= \frac{1}{\sigma^4}\left[ \sum_{x=1}^n x^3 \frac{n!}{(x-1)!(n-x)!} p^x q^{n-x} \right] - \frac{4\Sigma \sigma^3 \mu + 6\sigma^2 \mu^2 + \mu^4 + 3\sigma^4}{\sigma^4}\\
\end{align}
$$

This time, use $x^3 = 1 + 7(x-1) + 6(x-1)(x-2) + (x-1)(x-2)(x-3)$.  The first three terms are almost identical to what we found above.  And the fourth term is very similar:

$$
\begin{align}
\sum_{x=1}^n (x-1)(x-2)(x-3) \frac{n!}{(x-1)!(n-x)!} p^x q^{n-x} &= n(n-1)(n-2)(n-3)p^4 \sum_{x=4}^n \frac{(n-4)!}{(x-4)!(n-x)!} p^{x-4} q^{n-x} \\
&= n(n-1)(n-2)(n-3)p^4 \sum_{x=4}^n p_B(x-4| n-4, p) \\
&= n(n-1)(n-2)(n-3)p^4
\end{align}
$$

Putting these back into our original expression, we find

$$
\begin{align}
~K &= \frac{1}{\sigma^4}\left[ np + 7n(n-1)p^2 + 6n(n-1)(n-2)p^3 + n(n-1)(n-2)(n-3)p^4 \right] - \frac{4\Sigma \sigma^3 \mu + 6\sigma^2 \mu^2 + \mu^4 + 3\sigma^4}{\sigma^4}\\
&= \frac{np + 7n(n-1)p^2 + 6n(n-1)(n-2)p^3 + n(n-1)(n-2)(n-3)p^4 - 4n^2p^2(1-p)(1-2p) - 6n^3p^3(1-p) - n^4p^4 - 3n^2p^2(1-p)^2}{\sigma^4}\\
&= \frac{np(1-p) - 6np^2(1-p)^2}{n^2p^2(1-p)^2}\\
&= \boxed{\frac{1-6pq}{npq}}
\end{align}
$$


### END SOLUTION

**Question 3**: 

*Drawing  random  deviates  from  arbitrary  functions.* Although $\texttt{np.random}$ has numerous  functions from which you can draw random deviates, in real life you will encounter cases that are not in $\texttt{numpy}$.  This problem will guide you through the steps of writing your own function that samples from an arbitrary distribution.  Consider an asymmetric gaussian distribution function given by the following:

$$
f_{2G}(x| \sigma_L, \sigma_R) \propto
\begin{cases}
    e^{-x^2/2\sigma_L^2} & x < 0 \\
    e^{-x^2/2\sigma_R^2} & x \ge 0 \\
\end{cases}
$$

The function is continuous across $x=0$.
  
(a) (3 pts) Write a function that returns the value of $f_{2G}(x| \sigma_L, \sigma_R)$ normalized to unity, i.e., such that
  
$$
\int^\infty_{-\infty} f_{2G}(x| \sigma_L, \sigma_R) dx = 1
$$
  
The function should take three arguments: $\sigma_L$, $\sigma_R$, and $\texttt{x}$, a vector of $x$ values.

(b) (3 pts) Write a function that returns the cumulative distribution function ($CDF$) of the normalized $f_{2G}(x| \sigma_L, \sigma_R)$, i.e.,

$$
CDF(x; \sigma_L, \sigma_R) = \int^x_{-\infty} f_{2G}(x^\prime| \sigma_L, \sigma_R) dx^\prime
$$

This function should also take three arguments: $\sigma_L$, $\sigma_R$, and a vector of $x$ values.

(c) (4 pts) Using the results from above, write a function that draws random deviates from $f_{2G}$.  The sampler should take three arguments: $\sigma_L$, $\sigma_R$, and the number of random numbers to return.

(d) (2 pts) Draw $10^6$ random numbers from this distribution for $\sigma_L = 0.4$ and $\sigma_R = 1.2$ and plot a histogram of the values from $-2 \le x \le 5$.  Overplot the analytic function on the same graph.


```python
# Question 3a

import numpy as np

def f2gauss_pdf(sigma_L, sigma_R, x):
    """Return f_2G(x) for the given sigma_L, sigma_R and x values.
    """
    # Hints:
    # 1. Start with a numpy array the same shape as the input x values, but all zeros.
    #    You can use the numpy function np.zeros or np.zeros_like to do this.
    # 2. Numpy indexing will be helpful here.  e.g. pdf[x>=0] = 3 will assign all values
    #    in the pdf array corresponding to values with x>=0 to the value 3.  Obviously, you
    #    don't want to set the values to 3, but you can put the math you want on the rhs
    #    of this and it will only assign to those elements on the lhs where x>=0.
    #    Repeat this with pdf[x<0] for the other half.
    # 3. The slightly funny looking x[x>=0] will give you the x values that are >= 0.
    #    This will be the thing to use for x on the rhs.
    # 4. Start with the function normalized to have f_2G(0) = 1.
    #    You should be able to figure out the integral of this function from -inf..inf.
    #    Divide by this result to get a version where the integral is 1.

    ### BEGIN SOLUTION
    
    # Start with an empty array of the pdf.
    pdf = np.zeros_like(x)
    
    # Use numpy indexing by boolean array to do positive and negative halves separately.
    pdf[x<0] = np.exp(-x[x<0]**2 / (2. * sigma_L**2))
    pdf[x>=0] = np.exp(-x[x>=0]**2 / (2. * sigma_R**2))
    
    # This now is the function normalized to have pdf(0) = 1.
    # The integral of this from -inf..inf is sqrt(2pi) * (sigma_L + sigma_R)/2
    # So now divide by this value.
    pdf /= np.sqrt(2*np.pi) * (sigma_L + sigma_R)/2.
    
    return pdf
    ### END SOLUTION
```

```python
sigma_L = 2
sigma_R = 3

# Test with a few select values you may be able to check directly whether they make sense.
x_vals = np.array([-50., -4., -2., 0., 3., 6., 50.])

pdf = f2gauss_pdf(sigma_L, sigma_R, x_vals)
# If you haven't seen zip before, it basically lets you loop over two lists (or arrays) at once.
for x, p in zip(x_vals, pdf):
    print(f"pdf({x}) = {p}")

# Check the normalization using values that are dense enough to give a good estimate of the integral.
x_vals = np.linspace(-50, 50, 1000)
pdf = f2gauss_pdf(sigma_L, sigma_R, x_vals)

# Numpy has a handy function that implements the trapezoidal estimate of integration when you have
# both the x and y values as arrays.
integral = np.trapz(pdf, x_vals)
print('Integral = ',integral)
assert np.isclose(integral, 1.0)
### BEGIN HIDDEN TESTS

# This code snippet lets us just log the test failures so we can see all the things that fail, rather than
# bomb out on the first failure.  Then at the end, we just assert that there were 0 failures to get the
# automatic grading to give points.
nfail=0
from contextlib import contextmanager
import traceback
@contextmanager
def log_assert():
    global nfail
    try:
        yield
    except AssertionError as e:
        print('Failed assert:')
        print(traceback.format_exception(e)[-2].split('\n')[1])
        print('    msg =',str(e))
        nfail += 1

# Check a few other sigma values

for sL, sR in [(2.1,1.2), (0.3,0.5), (1,1), (2,3), (0.4,1.2)]:
    xv = np.arange(-10,10,dtype=float)
    pv = f2gauss_pdf(sL, sR, xv)
    #print(sL,sR,pv)
    with log_assert():
        assert len(pv) == len(xv), (len(pv), len(xv))
    # First check the unnormalized version.
    pv1 = pv / pv[10]
    with log_assert():
        assert np.allclose(pv1[xv>=0], np.exp(-xv[xv>=0]**2 / (2. * sR**2))), pv1[xv>=0]
    with log_assert():
        assert np.allclose(pv1[xv<=0], np.exp(-xv[xv<=0]**2 / (2. * sL**2))), pv[xv<=0]
    
    # Now check normalization.
    with log_assert():
        assert np.isclose(pv[10], 1. / (np.sqrt(2*np.pi) * (sL + sR)/2.)), pv[10]

print(f"\nTotal of {nfail} test failures")
assert nfail == 0
### END HIDDEN TESTS
```

```python
# Question 3b

def f2gauss_cdf(sigma_L, sigma_R, x):
    """Return a cumulative probability distribution for f_2G(x).
    
    i.e. int(f_2G(x'), x'=-infinity..x)

    You may assume that the input x values are linearly spaced and have sufficient 
    range and density to adequately sample the distribution, so that numerical 
    estimates of the integral using the given values are reasonably accurate.
    
    Returns the array cdf, the same size as x.
    """
    # Hints:
    # 1. Use f2gauss_pdf to get the pdf.
    # 2. The trapezoid rule gives a good approximation for the running integral.
    #    However, using np.trapz for each index would be extremely slow.
    #    So, think about how to use np.cumsum to compute the same thing faster.
    # 3. When you are done, cdf[0] should == 0 and cdf[-1] should == 1.
    
    ### BEGIN SOLUTION
    # The trapezoid rule sums up (pdf[i-1] + pdf[i])/2 for each interval
    # For a given index n, 
    #    cdf[n] = (sum_i=1..n (pdf[i-1] + pdf[i]) / 2.
    #           = sum_i=0..n-1 pdf[i]/2  +  sum_i=1..n pdf[i]/2

    # Get the pdf values from f2gauss_pdf
    pdf = f2gauss_pdf(sigma_L, sigma_R, x)
    
    # Start with an array of zeros the same size as pdf
    cdf = np.zeros_like(pdf)
    
    # This function computes the cumulative running sum of the given array.
    # cum[n] = sum_i=0..n pdf[i]
    cum = np.cumsum(pdf)
    
    # Leave cdf[0] == 0, and calculate the rest from cum.
    cdf[1:] = (cum[1:] + cum[:-1])/2.
    
    # cdf[-1] is now the full integral.
    # So divide the whole array by this to normalize.
    cdf /= cdf[-1]
    
    return cdf
    ### END SOLUTION
```

```python
# Problem 3c
def find_x_from_cdf(cdf, x, c):
    """Given a cdf(x) tabulated in numpy arrays cdf and x, and an array of 
    random numbers c, where 0 < c[i] < 1 for all i, find the x[i] values that 
    correspond to the given c[i] = cdf(x[i]).
    
    Returns a number array of x values corresponding to the given c values.
    """
    # Hints:
    # 1. The numpy function np.searchsorted will be helpful.
    # 2. You can get two values for cdf(x) that bracket each c value.
    # 3. Linearly interpolate between these values to get the corresponding x.

    ### BEGIN SOLUTION
    # We need to find the index k in the cdf array such that cdf[k-1] < c < cdf[k]
    # This will let us determine what x corresponds to this given cdf value.
    
    # The searchsorted function finds the index for each element in the cdf array 
    # such that cdf[k-1] < c < cdf[k]
    k = np.searchsorted(cdf, c)
    assert np.all(k > 0)         # This is only guaranteed if pdf[0] = 0
    assert np.all(k < len(cdf))  # This is only guaranteed if pdf[-1] = 0

    # c0 and c1 bracket each c value.
    c0 = cdf[k-1]
    c1 = cdf[k]
    assert np.all(c0 < c)  # Assert statements like these help when developing to make sure
    assert np.all(c <= c1) # the code is doing what you think it is doing.
    
    # x0 and x1 are the corresponding x values which bracket the value we want.
    x0 = x[k-1]
    x1 = x[k]
    
    # Linearly interpolate between (c0, x0) and (c1, x1)
    # m = (x1-x0) / (c1-c0)
    # x = m (c - c0) + x0
    return (x1-x0)*(c-c0)/(c1-c0) + x0
    ### END SOLUTION

def f2gauss(sigma_L, sigma_R, size):
    """Generate an array of samples drawn from the probability function 
    
        f_2G(x) = exp(-x^2/sigma_L^2) if x < 0
                  exp(-x^2/sigma_R^2) if x >= 0
                  
    Returns the array
    """
    # This function is relatively trivial now that you have done all the other functions.
    # 1. Select an appropriate range and density for the x values.
    # 2. Compute the corresponding cdf values with f2gauss_cdf.
    # 3. Create an array of uniform random deviates of length, size
    # 4. Use find_x_from_cdf to find x for each c value.

    ### BEGIN SOLUTION
    
    # 1. Select an appropriate range and density for the x values
    x = np.linspace(-20*sigma_L, 20*sigma_R, 1000)

    # 2. Compute the corresponding cdf values with f2gauss_cdf.
    cdf = f2gauss_cdf(sigma_L, sigma_R, x)
    
    # 3. Create an array of uniform random deviates of length, size
    # These are the cdf(x) values corresponding to the values x that we want.
    c = np.random.random(size)

    # 4. Use find_x_from_cdf to find x for each c value.
    x = find_x_from_cdf(cdf, x, c)
    
    return x

    ### END SOLUTION
```

```python
sigma_L = 0.4
sigma_R = 1.2
size = int(1.e6)

values = f2gauss(sigma_L, sigma_R, size)

print('Mean, var = {:.4f}, {:.4f}'.format(values.mean(), values.var()))
### BEGIN HIDDEN TESTS
nfail = 0

# This calculation is analytic in this case.  An exercise left to the reader.
true_mean = (2./np.pi)**0.5 * (sigma_R-sigma_L)
true_var = sigma_R*sigma_L + (sigma_R-sigma_L)**2 - true_mean**2
print('\nAnalytic values of mean, var: {:.4f}, {:.4f}'.format(true_mean, true_var))

# First some simple tests of the values array
with log_assert():
    assert len(values) == size, (len(values), size)
with log_assert():
    assert np.all(values > -10*sigma_L), values
with log_assert():
    assert np.all(values < 10*sigma_R), values

# Check find_x_from_cdf
# There are different choices possible for how to interpolate between the tabulated x values.
# So just check that x is in the right range.
xv = np.linspace(-50, 50, 1000)
cdf = f2gauss_cdf(sigma_L, sigma_R, xv)
c1 = np.random.random(5)
x1 = find_x_from_cdf(cdf, xv, c1)
for xx, cc in zip(x1,c1):
    i = np.searchsorted(xv, xx)
    with log_assert():
        assert xv[i-1] <= xx <= xv[i], (xv[i-1], xx, xv[i])
    with log_assert():
        assert cdf[i-1] <= cc <= cdf[i], (cdf[i-1], cc, cdf[i])

# Finally check that the actual values array computed has the right mean, var.
with log_assert():
    assert np.isclose(values.mean(), true_mean, rtol=1.e-2), (values.mean(), true_mean)
with log_assert():
    assert np.isclose(values.var(), true_var, rtol=1.e-2), (values.var(), true_var)

# Check some other sigma values:
for sL, sR in [(0.3,0.5), (1,1), (200, 300)]:
    print('sL,sR = ',sL,sR)
    v = f2gauss(sL, sR, size)
    true_mean = (2./np.pi)**0.5 * (sR-sL)
    true_var = sR*sL + (sR-sL)**2 - true_mean**2
    print('mean = ',v.mean(),true_mean)
    print('var = ',v.var(),true_var)
    with log_assert():
        assert np.isclose(v.mean(), true_mean, rtol=1.e-2, atol=1.e-2), (v.mean(), true_mean)
    with log_assert():
        assert np.isclose(v.var(), true_var, rtol=1.e-2), (v.var(), true_var)

print(f"\nTotal of {nfail} test failures")
assert nfail == 0
### END HIDDEN TESTS
```
